```python
print(data.info())
print('\n')
print(data.dtypes)
```

```python
#Removing unnecessary columns
    
data=data[['OrderDate', 'Region', 'City', 'Category', 'Product', 'Quantity',
       'UnitPrice', 'TotalPrice', 'Week', 'Month', 'Month Name', 'Day Name']]
print(data.shape)
data.head(7)
```

```python
#Count
print(data['Product'].value_counts())
print('\n')
print(data['Product'].unique())
```

```python
#Smallest, Largest
data.nsmallest(5,'UnitPrice') 
data.nlargest(5,'TotalPrice') 
```

```python
#groupby
data.groupby(by=['Product','City']).sum()[['Quantity','TotalPrice']]
        #data.groupby(by=['City','Product']).sum() -- Shows all the numerical columns
```

```python
#new column

data['10%dis']=data['TotalPrice']-(0.1*data['TotalPrice'])
print(data.shape)
data.head(7)
```

```python
#Delete column
del data['10%dis'] 
print(data.shape)
data.head(7)
```

```python
#Copying a column to a new column
data['copied']=data['City']+'-'+data['Product']
print(data.shape)
data.head(7)
```

```python
#Replacement
data['City']=data['City'].replace('Los Angeles','LA')
data['City']=data['City'].replace('New York','NY')
print(data.shape)
data.head(7)
```

```python
#Percentage and rounding 
data['%']=(data['TotalPrice']/data['TotalPrice'].sum())*100
data['%']=round(data['%'],2)
print(data.shape)
data.head(7)
```

```python
#Filter
NY_data=data[(data['City']=='NY') & (data['Category']=='Cookies')]
print(data.shape)
NY_data
```

```python
#.loc function
zz = data.loc[ :5, ['City', 'Category','TotalPrice']]
zz
```

```python
#Sorting 
NY_data=NY_data.sort_values('TotalPrice', ascending=False)
NY_data
```

```python
#if-else condition with numpy

import numpy as np
data['sales'] = np.where(data['Quantity'] > 199, '200+', np.where(data['Quantity'] > 99, '100-200', '<100'))
data.head(7)  
```

```python
#.loc function again

data.loc[data['TotalPrice'] > 300,'sale_made']='300+'
data.loc[data['TotalPrice'] > 100,'sale_made']='100-300'
data.loc[data['TotalPrice']<= 100,'sale_made']='<100'
data.head(7)  
```

```python
#Coverting to other data type
data['OrderDate'] = data['OrderDate'].astype(str)
data.head(7)
```

```python
#Pivot_Table 1

z=pd.pivot_table(data,index='City',values='TotalPrice',aggfunc=['sum','mean','max'])
round(z,2)
```

```python
#Pivot_Table 2

z=pd.pivot_table(data,index=('Category','Product'), columns='City', values=('TotalPrice','Quantity'),aggfunc='sum',margins=True)
round(z,2)
```
